```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.info()
```

```python
sns.boxplot(df['minimum_nights'])
df['minimum_nights'].describe()
```

```python
sns.distplot(df[df['price']<2000]['price'])
```

```python
sns.boxplot(df['price'])
```

```python
sns.violinplot(df[df['price']<2000]['price'])
```

```python
sns.countplot(data = df, x = 'neighbourhood_group', hue = 'room_type')
```

```python
df2 = sns.load_dataset('titanic')
df2.head()
```

```python
sns.countplot(data = df2, x = 'survived',hue = 'sex')
```

```python
sns.countplot(data = df2, x = 'survived',hue = 'class', palette = 'rainbow')
```

```python
sns.countplot(data = df2, x = 'class')
```

```python
sns.barplot(data= df2, x = 'class', y = 'age', hue= 'sex')
```

```python
sns.barplot(data = df2, x ='class', y = 'fare', hue = 'sex')
```

```python
sns.boxplot(data = df2, x= 'class', y = 'age')
```

```python
sns.violinplot(data = df2, x= 'class', y = 'age', hue = 'sex')
```

```python
sns.distplot(df2[df2['class']=='First']['age'])
```

```python
#Bi-Variate Analysis using Categorical Coulmns
#creating a pivot table to create a heatmap
x = pd.crosstab(df2['survived'], df2['class'])
sns.heatmap(x, annot = True, cmap='Blues', fmt = '.1f' )
```

```python
sns.stripplot(data = df2, x = 'class', y = 'age', hue = 'sex')
#this is with both numerical and categorical
```

```python
df3 = sns.load_dataset('tips')
df3.head()
```

```python
#just numerical data
sns.scatterplot(x = 'total_bill', y = 'tip', data = df3)
```

```python
#you can also analyse numerical and categorical data
sns.scatterplot(data = df3, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker')
```

```python
#analysing tips and bills in gender
sns.scatterplot(x = 'total_bill', y = 'tip', data = df3, hue = 'sex')
```

```python
sns.jointplot(data = df3, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df3, hue = 'sex')
```

```python
#Exploring Line Plot
df4 = sns.load_dataset('dowjones')
df4.head()
#df4.shape
```
